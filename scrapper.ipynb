```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests

    
```

```python
#configurations 
headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }

#Scrapping
url='https://www.riyasewana.com/search/motorcycles'
response = requests.request("GET", url, headers=headers)

# parse the downloaded data
data = bs(response.content, 'html.parser')
#cards=data.find_all('div',attrs={'id','wrapper'})
#print(len(cards))
cards=data.find_all('li',attrs={'class','item round'})
```

```python
scraped_data=[]
img_url=[]

for card in cards[:15]:
    Bike_name_year=card.find('h2').text
    Bike_image=card.find('img')['src']
    Bike_city=card.find('div',attrs={'class','boxintxt'}).text[1::]
    Bike_price=card.find('div',attrs={'class','boxintxt b'}).text[5::]
    Date_posted=card.find('div',attrs={'class','boxintxt s'}).text[1::]
    if Bike_price!='tiable':
        Bike_price=int(Bike_price.replace(',',''))
    else:
        continue
    card_details={}
    card_details['Bike_name_year']=Bike_name_year
    card_details['Bike_image']=Bike_image
    card_details['Bike_price']=Bike_price
    card_details['Bike_city']=Bike_city
    card_details['Date_posted']=Date_posted
    response=requests.get('http:'+Bike_image)
    
    # Save the images into or folder
    
    with open(Bike_name_year+'.jpg', 'wb') as handle:
   
        for block in response.iter_content(1024):
            if not block:
                break

            handle.write(block)
    
    scraped_data.append(card_details)
```

```python
bikes=pd.DataFrame.from_dict(scraped_data)
bikes.head(10)
```
